```python
def remove_outlier(co1):
    sorted(co1)
    Q1,Q3=co1.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5+IQR)
    upper_range=Q3+(1.5+IQR)
    return lower_range,upper_range
```

```python
lowstays_in_weekend_nights,uppstays_in_weekend_nights=remove_outlier(mydata['stays_in_weekend_nights'])
mydata['stays_in_weekend_nights']=np.where(mydata['stays_in_weekend_nights']>uppstays_in_weekend_nights,uppstays_in_weekend_nights,mydata['stays_in_weekend_nights'])
mydata['stays_in_weekend_nights']=np.where(mydata['stays_in_weekend_nights']<lowstays_in_weekend_nights,lowstays_in_weekend_nights,mydata['stays_in_weekend_nights'])
```
