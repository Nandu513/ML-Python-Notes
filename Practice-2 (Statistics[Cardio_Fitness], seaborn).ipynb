## Statistics

```python
#measures of central tendency(mean,median,mode)
#mean can be impacted by outlier
#no mean for categorical data/qualitative data
#if there missing vales & no outliers we can replace it by mean
#all observations are used
```

```python
#median:middle most 
#arrange in asc/des
#not much impacted by missing values,outlier
#no median for categorical data
#all observations are not used
```

```python
#mode:frequency
#can use for categorial data
#no outlier missing values can be replaced by mean/median
#when outlier present use mean
#not affrcted by outlier
#all obersations are not used
```

```python
#gausian distribution curve/bell curve: mean=median=mode
#skewed curve: mean!=median!=mode
```

```python
#measures of dispersion(range,variance,standard deviation)
```
